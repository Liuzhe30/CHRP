```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_pickle('../dataset/merged_data.pkl')
data.head()
```

```python
con_list = ['n342','n346','n353','n366','n369','n431','n438']
marker_list = ['AL355516.1','CPEB2-DT']
```

```python
expression = np.zeros((2,7),dtype=float)
for i in range(7):
    expression[0][i] = data.loc[con_list[i]+'c',marker_list[0]] - data.loc[con_list[i],marker_list[0]]
    expression[1][i] = data.loc[con_list[i]+'c',marker_list[1]] - data.loc[con_list[i],marker_list[1]]
print(expression)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(expression,cmap='coolwarm',annot=True,xticklabels=con_list,yticklabels=marker_list,vmin=-0.3)
plt.title('2-year Changes of RNA marker expression\n of CHR-P conversion individuals')
plt.savefig('../images/expression-2year.png',dpi=300)
plt.show()
```
