```python
import pandas as pd
import numpy as np
```

```python
def normalization(data):
    _range = np.max(np.max(abs(data)))
    return data / _range
```

```python
#filtered.to_csv('../dataset/meth_probe.csv',index='probeID')
filtered = pd.read_csv('../dataset/meth_probe.csv',index_col=0)
filtered.head()
```

```python
filtered = normalization(filtered)
filtered.head()
```

```python
merged = pd.read_pickle('../dataset/2year_merged_data.pkl')
merged = merged[['end']]
end = merged[['end']]
merged = merged[merged.index.isin(filtered.columns.values.tolist())]
print(merged.shape)
merged
```

```python
cg = filtered.T
new_merged = merged.join(cg)
print(new_merged.shape)
new_merged
```

```python
new_merged.to_pickle('../dataset/2year_meth_merged_dataset.pkl')
```
