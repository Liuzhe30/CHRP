#### preprocessing 2-year conversion dataset

```python
import pandas as pd
import numpy as np
```

```python
# load dataset
anno = pd.read_csv('../dataset/annofix-2year.csv',index_col=0)
edata = pd.read_csv('../dataset/final_TPM.csv',index_col=0)
```

```python
anno.head()
```

```python
edata = edata.T
edata
```

```python
merged = anno.join(edata)
merged
```

```python
merged = merged.loc[:, ~merged.columns.duplicated()]

print(merged)
merged.to_pickle('../dataset/2year_merged_data.pkl')
```

```python
merged = pd.read_pickle('../dataset/2year_merged_data.pkl')
merged.replace('f', 1, inplace=True)
merged.replace('m', 0, inplace=True)
merged.replace('re', 0, inplace=True)
merged.replace('sym', 0, inplace=True)
merged.replace('con', 1, inplace=True)
merged
```

```python
label = np.array(merged['end'])
label
```

```python
merged.to_pickle('../dataset/2year_merged_data_2class.pkl')
```
