```python
import pandas as pd
import numpy as np

def normalization(data):
    _range = np.max(np.max(abs(data)))
    return data / _range
```

```python
label = pd.read_csv('../dataset/twomics_phen_group.csv')
label
```

```python
filtered_dnam = pd.read_csv('../dataset/meth_probe.csv',index_col=0)
filtered_dnam.head()
filtered_dnam = normalization(filtered_dnam)
filtered_dnam.head()
```

```python
edata = pd.read_csv('../dataset/final_TPM.csv',index_col=0)
edata = normalization(edata)
edata.head()
```

```python
filtered_dnam_T = filtered_dnam.T
edata_T = edata.T
filtered_dnam_T = filtered_dnam_T.reset_index().rename(columns={"index": "sample"})
edata_T = edata_T.reset_index().rename(columns={"index": "sample"})
merged_df = label.merge(filtered_dnam_T, on="sample", how="left")
merged_df = merged_df.merge(edata_T, on="sample", how="left")
merged_df = merged_df.drop(columns=['group'])
merged_df.to_pickle('../dataset/rna-dnam-merge-raw.pkl')
merged_df.head()
```

```python
merged_df = pd.read_pickle('../dataset/rna-dnam-merge-raw.pkl')
merged_df.replace('re', 1, inplace=True)
merged_df.replace('sym', 0, inplace=True)
merged_df.replace('con', 0, inplace=True)
merged_df.to_pickle('../dataset/rna-dnam-merge-2class.pkl')
merged_df.head()
```

```python
merged_df = pd.read_pickle('../dataset/rna-dnam-merge-raw.pkl')
merged_df.replace('re', 0, inplace=True)
merged_df.replace('sym', 1, inplace=True)
merged_df.replace('con', 2, inplace=True)
merged_df.to_pickle('../dataset/rna-dnam-merge-3class.pkl')
merged_df.head()
```
