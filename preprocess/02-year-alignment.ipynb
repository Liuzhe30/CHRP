```python
import pandas as pd
import numpy as np
```

```python
merged = pd.read_pickle('../dataset/merged_data.pkl')
merged
```

```python
index_list = merged.index.to_list()
new_list = []
new_list2 = []
for i in index_list:
    if(i+'c' in index_list):
        new_list.append(i)
        new_list2.append(i+'c')
print(new_list2)
```

```python
new_dataset = merged[merged.index.isin(new_list2)]
new_dataset = new_dataset[['AC026691.1','CA8','end']]
year1f1 = merged[merged.index.isin(new_list)]['AC026691.1'].tolist()
year1f2 = merged[merged.index.isin(new_list)]['CA8'].tolist()
year1label = merged[merged.index.isin(new_list)]['end'].tolist()
print(year1label)
new_dataset.insert(new_dataset.shape[1],'year1-AC026691.1',year1f1)
new_dataset.insert(new_dataset.shape[1],'year1-CA8',year1f2)
new_dataset.insert(new_dataset.shape[1],'year1-label',year1label)
new_dataset.to_pickle('../dataset/year2_merged_dataset.pkl')
new_dataset
```
